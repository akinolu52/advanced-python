```python
import cv2
import pandas as pd
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('plane.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# to get the image datatype
type(image)
```

```python
# to get the number of rows and columns
rows, columns = image.shape

rows, columns
```

```python
# convert to a dataframe
df = pd.DataFrame(image)

df.head()
```

```python
image_2 = image[int(rows / 2):, int(columns / 2):]
```

```python
# cropping

plt.imshow(image_2, cmap='gray'), plt.axis('off')
plt.show()
```

```python
image_bgr = cv2.imread('./plane.jpg', cv2.IMREAD_COLOR)
```

```python
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb), plt.axis('off')
plt.show()
```

```python
plt.imshow(image_bgr), plt.axis('off')
plt.show()
```

### Resize image by changing the dimension of the image

```python
image_50x50 = cv2.resize(image, (50, 50))
plt.imshow(image_50x50), plt.axis('off')
plt.show()
```

```python
image_50x50.shape
```

### Cropping Images

```python
image_cropped = image[:int(rows / 2), :]
plt.imshow(image_cropped), plt.axis('off')
plt.show()
```

### Saving Images

```python
cv2.imwrite('./plane_50x50.png', image_50x50)
```
